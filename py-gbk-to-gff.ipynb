```python
from pathlib import Path
from Bio import SeqIO
from gfflu.io import write_gff
```

```python
gbk = Path('phold.gbk')
```

```python
recs = list(SeqIO.parse(gbk, 'genbank'))
```

```python
len(recs)
```

```python
from gfflu.io import flatten_features
```

```python
rec = recs[0]
rec = flatten_features(rec)
```

```python
rec.features
```

```python
write_gff([rec], 'NC_044954.1.gff')
```
